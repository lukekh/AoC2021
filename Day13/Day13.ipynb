<p>The transparent paper is pretty big, so for now, focus on just completing the first fold. After the first fold in the example above, <code><em>17</em></code> dots are visible - dots that end up overlapping after the fold is completed count as a single dot.</p>
<p><em>How many dots are visible after completing just the first fold instruction on your transparent paper?</em></p>

```python
# Initialise coordinates and instructions from text
with open('Day13.in') as f:
    coordinates, instructions = f.read().split('\n\n')
    
    # Cleaning helper functions
    clean_coord = lambda x: tuple(map(int, x.split(',')))
    clean_instruction = lambda along, y: (along[-1], int(y))
    
    # Final cleaned data
    coordinates = set([clean_coord(c) for c in coordinates.split('\n')])
    instructions = [clean_instruction(*i.split('=')) for i in instructions[:-1].split('\n')]
```

```python
# Folding functions along axes
fold_x = lambda along, x, y: (along - abs(x-along), y)
fold_y = lambda along, x, y: (x, along - abs(y-along))

def fold(var, along, paper):
    """
    Take a var (x/y) and the line along which to fold and then iterate over
    the coordinates in paper and return new folded sheet
    """
    if var == 'x':
        return set([fold_x(along, *p) for p in paper])
    elif var == 'y':
        return set([fold_y(along, *p) for p in paper])

# Fold once
folded = fold(*instructions[0], coordinates)
    
# Solution
print(f"There are {len(folded)} points after the first fold")
```

## Part Two

<p><span title="How can you fold it that many times? You tell me, I'm not the one folding it.">Finish folding</span> the transparent paper according to the instructions. The manual says the code is always <em>eight capital letters</em>.</p>
<p><em>What code do you use to activate the infrared thermal imaging camera system?</em></p>

---

```python
# Continue on from part one
for i in instructions[1:]:
    folded = fold(*i, folded)

# Initialise string and size of display
display = ''
x_edge, y_edge = max(folded, key=lambda x:x[0])[0]+1, max(folded, key=lambda x:x[1])[1]+1

# Read print out
for y in range(y_edge):
    for x in range(x_edge):
        display += '▓' if (x,y) in folded else ' '
    display += '\n'

# Solution
print(f"Solution:\n{'*'*x_edge}\n\n{display}\n{'*'*x_edge}")
```
