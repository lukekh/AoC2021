<p>In this example, the sea cucumbers stop moving after <code><em>58</em></code> steps.</p>
<p>Find somewhere safe to land your submarine. <em>What is the first step on which no sea cucumbers move?</em></p>

---

```python
# Initialise
with open('Day25.in') as f:
    inputs = [i[:-1] for i in f.readlines()]
    Y, X = len(inputs), len(inputs[0])
    south = set()
    east = set()
    for y, row in enumerate(reversed(inputs)):
        for x, char in enumerate(row):
            if char == '>':
                east.add((x,y))
            elif char == 'v':
                south.add((x,y))
```

```python
# Helper functions
sstep = lambda x,y, Y=Y: (x, (y-1) % Y)
estep = lambda x,y, X=X: ((x+1) % X, y)

def steps(i=0, south=south, east=east, X=X, Y=Y):
    """
    step all east-herd sea-cucumbers, then step all south-herd sea-cucumbers
    """
    U = south | east
    new_e = set([e if (n:=estep(*e)) in U else n for e in east])
    
    U = south | new_e
    new_s = set([s if (n:=sstep(*s)) in U else n for s in south])
    
    if (new_s!=south) or (new_e!=east):
        return steps(i+1, south=new_s, east=new_e, X=X, Y=Y)
    else:
        return i
    
    
# Solution
print(f"After {steps()} steps, no sea-cucumbers move.")
```
